## <b>EEG Glossary</b>

1. <b>Boundary Element Method (BEM)</b>:

    Boundary Element Method (BEM) is a numerical method that simulates the electric potential and magnetic field produced outside the brain by neural activity.


2. <b>Power spectral density (PSD)</b>:
    - Power Spectral Density (PSD) tells you basically how power is distributed over the frequency range.  
    - It is the average power a signal carries in 1 Hz of bandwidth. 


3. <b>Alpha Power</b>:

    Alpha power refers to the magnitude or strength of brain activity within the alpha frequency range, typically measured in microvolts squared (µV^2) or decibels (dB). In EEG (electroencephalography) studies, alpha power is computed by analyzing the amplitude or power of neural oscillations specifically within the alpha frequency band, which typically ranges from 8 to 13 cycles per second (Hertz).

5. <b>Sampling-Rate</b>:

    In EEG (Electroencephalography) recordings, the sampling rate refers to how frequently the electrical signals from the brain are measured per second. A higher sampling rate means that more data points are collected per second, providing a finer resolution of the brain's electrical activity over time.

    For instance, when EEG signals are recorded at a sampling rate of 1000 Hz (Hertz), it means that the electrical activity of the brain is being measured 1000 times per second. This high sampling rate allows researchers to capture very fast changes in brain activity with great precision.

    Recording EEG signals at high sampling rates, such as 1000 Hz or higher, is often necessary when studying brain dynamics that occur rapidly, such as event-related potentials (ERPs) or transient brain responses to stimuli. It ensures that no critical information is missed and provides detailed insights into the temporal dynamics of brain processes.

    However, it's important to note that recording at high sampling rates also increases the amount of data collected and requires more memory and computational resources for storage and analysis. Therefore, the choice of sampling rate should be balanced based on the specific requirements of the study and the available resources.


6. <b>Epochs</b>:

    - In EEG (Electroencephalography) data analysis, epochs refer to equal-length segments of data that are extracted from continuous EEG recordings.
    - Usually, epochs are extracted around stimulus events or responses, but sometimes sequential or overlapping epochs are used.


7. <b>Independent Component Analysis (ICA)</b>:
    - Independent Component Analysis is a technique used to separate mixed signals into their independent sources. 
    - Independent Component Analysis (ICA) is a statistical and computational technique used in machine learning to separate a multivariate signal into its independent non-Gaussian components.
    - The heart of ICA lies in the principle of statistical independence. ICA identify components within mixed signals that are statistically independent of each other.


8. <b>Filtering</b>: 
    In signal processing, including EEG data analysis, filtering is a technique used to modify or isolate specific frequency components of a signal.
    - <b>High Pass Filtering</b>: 
        - High-pass filtering is a type of filtering that attenuates frequencies below a certain cutoff frequency while leaving higher frequencies unchanged. 
        - This type of filtering is designed to reduce or eliminate low-frequency components from the signal. It allows higher-frequency components to pass through without significant alteration.

    - <b>Low-Pass Filtering</b>:
        - Low-pass filtering is essentially the opposite of high-pass filtering. 
        - Instead of attenuating parts of the signal below a certain frequency, it attenuates parts of the signal above a certain frequency. 
        - It’s called low-pass because it lets frequencies lower than a certain cutoff pass.  


9. <b>Events</b>: 
    - Events in MNE provide a mapping between specific times during an EEG/MEG recording and what happened at those times.
    - Events are stored as a 2-dimensional NumPy array.
    - In MNE, there are two ways to create events: 
        - reading from a file/Raw object, or 
            - <i>mne.find_events(raw)</i>, here 'raw' is Raw object.
        - creating equal-sized events.
            - <i>mne.make_fixed_length_events(raw, start=0, stop=10, duration=1.)</i>, it creates second-long events for the first 10 seconds of EEG data.
            - <i>mne.make_fixed_length_events(raw, start=0, stop=10, duration=1., <b>overlap=0.5</b>)</i>, create events that overlap by a certain duration.
